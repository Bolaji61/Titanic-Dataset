```python
from sklearn.preprocessing import LabelEncoder
```

```python
gle = LabelEncoder()
train['Sex'] = gle.fit_transform(train['Sex'])
train.head()
```

```python
train = train[['PassengerId', 'Pclass', 'Sex', 'Fare', 'Survived']]
train.head()
```

```python
features = ['Pclass', 'Fare', 'Sex']
X = train[features]
Y = train['Survived']

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, train_size = 75, random_state = 0)
train_Y
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(train_X, train_Y)
X_predict = model.predict(test_X)
X_predict
```

```python
score = model.score(test_X, test_Y)
score
```

```python
X_predict = pd.DataFrame(X_predict)
X_predict.to_csv('result.csv')
```
